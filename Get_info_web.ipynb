```python
import pandas as pd
import nmap
import os
nm = nmap.PortScanner()
```

## The next two functions provide a DataFrame with the Scanned Information of the different ports from a web
###

```python
def get_info_web(web):
    
    """
    Returns a DataFrame with all the information extracted from the Given Web + IP, ports and TTL
    INPUT Web Domain or IP
    OUTPUT Clear DataFrame
    """
    
    scaneo = nm.scan(web)
    ports = scaneo['scan'][list(scaneo['scan'])[0]]['tcp']
    ip_web = list(scaneo['scan'].keys())[0]
    try:
        ttl = int(os.popen("ping -c 1 " + web).read().split("ttl=")[1].split(" ")[0])
    except:
        ttl = "Not received TTL"

    df = pd.DataFrame() 

    for port in list(ports.keys()):
        aux = pd.DataFrame({k: [v] for k , v in ports[port].items()})
        aux['port'] = port
        df = df.append(aux).reset_index(drop = True)

    df['web'] = scaneo['scan'][list(scaneo['scan'].keys())[0]]['hostnames'][0]['name']
    df['ip'] = ip_web
    df['ttl'] = ttl  

    return df

def get_info_lista_webs(lista_webs):

    """
    Returns a DataFrame with the information of a list of webs
    INPUT: List of domains to check or List or IPs
    OUTPUT: Clear DataFrame
    
    """
    df = pd.DataFrame()
    
    for web in lista_webs:
        try:
            df_aux = get_info_web(web)
            df = df.append(df_aux).reset_index(drop = True)
            print(web + ' done!')
        except:
            print(web + ' has not been possible to get access!')
            continue
        
    return df


        
```

```python
def get_ip(domain):
    
    """
    
    Returns for a given Domain, its IP
    INPUT: IP or Domain
    OUTPUR: list [Domain, IP]
    
    """
    info =  [e.replace(':\t', ' ') for e in os.popen("nslookup " + domain).read().split('\n')[2:] if 'Name' in e or 'Add' in e]
    return [e.replace('Name ', '').replace('Address: ', '') for e in info][1]
    
    
```
